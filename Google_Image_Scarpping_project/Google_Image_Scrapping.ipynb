```python
Soup =BeautifulSoup(response.content,'html.parser')
Soup
```

```python
images_tag = Soup.find_all("img")
images_tag
len(images_tag)
```

```python
del images_tag[0]
```

```python
img_data_mongo = []

for i in images_tag:
  image_url = i['src']
  image_data = requests.get(image_url).content
  my_dict = {"index":image_url,"image":image_data}
  img_data_mongo.append(my_dict)

  with open(os.path.join(save_dir,f"{query}_{images_tag.index(i)}.jpg"),"wb") as f:
    f.write(image_data)
```
